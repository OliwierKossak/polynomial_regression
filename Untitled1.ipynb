```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

# import data from csv file
data_df = pd.read_csv('cube_4x4_time.csv')

# stacking time is transformed from minutes:seconds:milliseconds format to seconds
time_list = []
for time in data_df['stacking_time']:
    time_list.append(time.split('.'))

time_in_seconds = []
for time in time_list:
    new_time = float(time[0]) * 60 + float(time[1]) + float(time[2]) * 0.01
    time_in_seconds.append(new_time)

data_df['stacking_time'] = time_in_seconds

# creates array with independent and dependent data
X = np.array(data_df['number_of_solution']).reshape(-1, 1)
y = np.array(data_df['stacking_time']).reshape(-1, 1)

# split data to train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Initialization of second degree of polynomial regression
poly = PolynomialFeatures(degree=2)

# training of polynomial regression
x_new = poly.fit_transform(X_train)
x_test_new = poly.fit_transform(X_test)

# training of linear regression
model = LinearRegression()
model.fit(x_new, y_train)

y_pred = model.predict(x_test_new)

#  allow you to fit a polynomial function to data
mymodel = np.poly1d(np.polyfit(np.array(data_df['number_of_solution']), np.array(data_df['stacking_time']), 2))

# Return evenly spaced numbers over a specified interval.
myline = np.linspace(2, 110, 130)


plt.scatter(X, y, color='cyan')
plt.ylabel('time in seconds')
plt.xlabel('number of solving the 4x4 cube')
plt.plot(myline, mymodel(myline))
plt.show()



```

```python
print('MAE training data: %.3f, test data: %.3f' % (
    mean_absolute_error(y_train, model.predict(x_new)),
    mean_absolute_error(y_test, model.predict(x_test_new))))

print('MSE training data: %.3f, test data: %.3f' % (
    mean_squared_error(y_train, model.predict(x_new)),
    mean_squared_error(y_test, model.predict(x_test_new))))

print('R^2 training data: %.3f, test data: %.3f' % (
    r2_score(y_train, model.predict(x_new)),
    r2_score(y_test, model.predict(x_test_new))))
```
